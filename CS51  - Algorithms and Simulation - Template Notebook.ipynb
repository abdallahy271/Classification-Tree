```python
# Imports the numpy array module
import numpy as np

t0 = 0 # stores the initial t(time) value 
t_end = 48 #stores the t values after which the code will stop

h = .1 #step size
steps = int((t_end - t0)/h + 1) # finds number of steps with the Euler formula 

t = np.linspace(t0, t_end, steps) # stores t values
S = np.zeros(steps) #  stores S values
I = np.zeros(steps) #  stores I values
R = np.zeros(steps) #  stores R values
N = 4535024  #the total population in the system
b =  0.050065 #the rate of infection of Corona
k = 0.0181   # Corona's rate of recovery

def dSdt(t,S,I):
    # defines the dS/dt and returns the change over time using the SIR formula
    return -b*S*I/N 

def dIdt(t,S,I):
    # dI/dt
    return (b*S/N-k)*I

def dRdt(t,S,I):
    #dR/dt
    return k*I

# sets the initial conditions
S[0] = 4365000 # Worldmeter(2020)
I[0] = 106211 # Worldmeter(2020)
R[0] = 63813 #Worldmeter(2020)

for i in range(steps-1): # runs a loop for the numbe4 of steps it takes
    S[i+1] = S[i] + dSdt(t[i],S[i],I[i])# Adds previous y-value of susceptible to change 
    I[i+1] = I[i] + dIdt(t[i],S[i],I[i])# Adds previous y-value of infected to change
    R[i+1] = R[i] + dRdt(t[i], S[i], I[i]) #Adds previous y-value of recovered to change

import matplotlib.pyplot as plt  #imports the visualization package
%matplotlib inline
#sets a fontsize
plt.rcParams.update({'font.size': 15}) 
plt.rcParams["figure.figsize"] = [8,5]
#plots the S,I and R nd labels the legends
plt.plot(t,S,linewidth=2,label='S(t)')
plt.plot(t,I,linewidth=2,label='I(t)')
plt.plot(t,R,linewidth=2,label='R(t)')
plt.xlabel('t (days)')
plt.ylabel('S,  I (people)')
plt.title("Coronavirus simulation using the Euler's method" )
plt.legend(loc='best')
plt.show()
```
